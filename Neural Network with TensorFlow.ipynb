##**Pengenalan TensorFlow**

**Model Neural Networks Dasar**
TensorFlow adalah sebuah library pada python untuk Deep Learning. Pada bagian ini teman-teman akan fokus pada syntax bagaimana melakukan pemodelan Neural Netwok menggunakan library TensorFlow python. Pada bagian ini kita akan mulai pada kasus yang paling sederhana yaitu Tabel Logika "And" sebagai berikut;

$x_1$  |$x_2$   |y
-------|------- |-----
1  | 1 | 1
1  | 0 | 0
0  | 1 | 0
0  | 0 | 0

Perhatikan bahwa x1 dan x2 adalah input dan y merupakan output yang diinginkan. Maka **ukiran input adalah dua kolom dan 4 baris sedangkan input adalah sebuah nilai antara 0 dan 1**.



```python
# Import Library
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

```python
# Menampilkan versi tensorflow
tf.__version__
```

Tahap Penting ketika ingin melakukan pemodelan Neural Network;
1. Menyusun arsitektur model
2. Mengkompilasi Model
3. Training model
4. Evaluasi model


```python
# Menyusun model
model=keras.Sequential([                            #Mmembentuk model sequntial
    tf.keras.layers.Dense(units=1,
                          input_shape=[2],
                          activation='relu'),       #Menambah layer pada model NN                       
])

# Compile model
model.compile(loss='mean_squared_error',            #menentukan fungsi loss
              optimizer='sgd')                      #Menentukan fungsi optimizer
model.summary()                                     #Menampilkan summary model
```

Berdasarkan model summary diatas maka kita sedang embuat model NN dengan satu titik layer. Kemudian hasil parameternya adalah tiga. Itu berarti terdapat tiga parameter, Apa saja? Yups betul w1, w2 karena inputnya adalah x1, x2, serta b yaitu bias. Berikut adalah visualisasi NN modelnya;

Screenshot (101).png

```python
# Mendefinisikan Data Input(X) dan Output (Y)
x=np.array([[1,1],[1,0],[0,1],[0,0]],dtype="int")
y=np.array([[1],[0],[0],[0]],dtype="int")
```

```python
#Menampilan data x
x
```

```python
#Menampilkan data y
y
```

```python
#Melatih model (fitting/training)
history=model.fit(x,y,epochs=2000, verbose=2)
```

```python
#Prediksi Model
model.predict(x)
```

```python
#Membulatkan HAsil Permodelan
model.predict(x).round()
```

```python
#Menampilkan nilai bobot
model.get_weights()
```
